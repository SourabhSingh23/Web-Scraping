```python
import pandas as pd
import lxml
import requests
from bs4 import BeautifulSoup
```

```python
url = "https://www.nfl.com/standings/league/2019/REG"
```

```python
page = requests.get(url)
page
```

```python
soup = BeautifulSoup(page.text, 'lxml')
```

```python
table = soup.find('table', summary = 'Standings - Detailed View')
```

```python
table.find_all('th')
```

```python
headers = []

for i in table.find_all('th'):
    title = i.text.strip()
    headers.append(title)

headers
```

```python
df = pd.DataFrame(columns=headers)
df
```

```python
rows = table.find_all('tr')[1:]
```

```python
for j in rows:
    row_data = j.find_all('td')
    row = [tr.text for tr in row_data]
    length = len(df)
    df.loc[length] = row
```

```python
for j in rows:
    first_td = row.find_all('td')[0].find('div', class_ = 'd3-o-club-fullname').text.strip()
    row_data = j.find_all('td')[1:]
    row = [tr.text.strip() for td in row_data]
    row.insert(0,first_td)
    length = len(df)
    df.loc[length] = row
```
